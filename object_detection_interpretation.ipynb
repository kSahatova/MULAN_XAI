# Generate interpretations for the predictions

## import Explainer

```python
import torch
from interpretability.explainers import Explainer 
from captum._utils.models.linear_model import SkLearnLinearRegression
from interpretability.interpretable_models import DecisionTree, train_tree
```

```python
explainer = Explainer()

explanation_methods = ['GradientShap', 'LayerGradientShap', 'LayerActivation', 'Occlusion']
target = torch.tensor([1], device=torch.device('cuda:0'))
```

```python
def rpn_forward_fused(input):
    features = model.backbone(input)
    features, images = model.feature_fuse(features, input)
    proposals = model.rpn.head(features)
    cls_logits, bbox_reg = proposals
    output = cls_logits[0].sum(dim=(2, 3))
    return output
```

```python
# 'lesion' class indexed as 1

exp_methods_attributes = {}

for method in explanation_methods:
    print(f'Explanation method: {method}')
    exp_methods_attributes[method] = explainer.get_explanations(method, images, rpn_forward_fused, 
                                                                model=model, target=target)
```

```python
lime_lr_attrs = explainer.get_explanations('LimeExplainer', images,
                                           rpn_forward_fused, target=target, kernel_width=0.1,
                                          interpretable_model=SkLearnLinearRegression())
```

```python
lime_dt_attrs = explainer.get_explanations('LimeExplainer', images,
                                           rpn_forward_fused, target=target, kernel_width=0.1,
                                          interpretable_model=DecisionTree(train_tree))
```

Blue colors represent more valuable attributes in the perturbation-based method, and red colors represent less valuable attributes.
Such a confusing result for Occlusion algorithm, for example, can be explained in the following way: dark patches that have been grayed out due to the grayscale pixels' intensities distribution are showed as more valuable for the prediction while bright patches representing inner tissues are highlighted as less significant. Patches that are already grayscale are less likely to influence prediction when perturbed with gray color patches, even if they contain useful features.
